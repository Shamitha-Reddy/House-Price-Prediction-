regression evaluation metrics

**Mean Absolute Error** (MAE) represents the average absolute deviation between the predicted values and the actual values. It's easy to comprehend, as it gives us the average magnitude of errors.

**Mean Squared Error** (MSE) is a widely adopted metric as it not only calculates the average error but also penalizes larger errors more significantly by squaring them. This property makes it particularly valuable in situations where we want to give greater importance to outliers or larger deviations from the true values.

**Root Mean Squared Error** (RMSE) takes the concept of MSE one step further by computing the square root of the average squared errors. RMSE provides an interpretable metric in the same units as the target variable, making it a popular choice for conveying the model's performance in a more relatable manner.

All of these metrics are considered **loss functions** because our goal is to minimize them. Lower values of MAE, MSE, or RMSE indicate better predictive accuracy in regression models.

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
